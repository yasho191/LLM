```python
# Load JSON data

def plot_accuracy(file_path, parameter, learning_rate):
    with open(file_path, 'r') as file:
        data = json.load(file)

    # Plotting training accuracies
    plt.figure(figsize=(15, 5))

    plt.subplot(1, 2, 1)
    for key, values in data.items():
        plt.plot(values[0], label=key)

    plt.title(f"Train Acc. for Different Model {parameter} | LR: {learning_rate}")
    plt.xlabel("Epoch")
    plt.ylabel("Accuracy")
    plt.legend(title=parameter)
    plt.grid(True)

    # Plotting validation accuracies
    plt.subplot(1, 2, 2)
    for key, values in data.items():
        plt.plot(values[1], label=key)

    plt.title(f"Val Acc. for Different Model {parameter} | LR: {learning_rate}")
    plt.xlabel("Epoch")
    plt.ylabel("Accuracy")
    plt.legend(title=parameter)
    plt.grid(True)
    plt.subplots_adjust(wspace=0.5)
    plt.tight_layout()
    plt.show()

    
plot_accuracy("results/0.001_model_depth_analysis.json", "Depth", "1e-3")
plot_accuracy("results/0.001_cls_hidden_dim_analysis.json", "Hidden Classifier Dim", "1e-3")
plot_accuracy("results/0.001_head_analysis.json", "Number of Heads", "1e-3")

plot_accuracy("results/0.003_model_depth_analysis.json", "Depth", "3e-3")
plot_accuracy("results/0.003_cls_hidden_dim_analysis.json", "Hidden Classifier Dim", "3e-3")
plot_accuracy("results/0.003_head_analysis.json", "Number of Heads", "3e-3")
```
